```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plt.show()
%matplotlib inline
```

```python
ageList = [10,20,30,40,54,63,74,85,89,91]
weightList = [20,45,67,60,55,45,78,68,68,53]
```

```python
numpyAgeList = np.array(ageList)
numpyWeightList = np.array(weightList)
```

## plot(x, y, "color")

```python
# draw graphic
plt.plot(numpyAgeList, numpyWeightList, "r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age - Weight Graphic")
```

```python
 numpyArrayList = np.linspace(0,10,20)
```

```python
numpyArrayList
```

```python
numpyArrayList2 = numpyArrayList ** 3 
```

```python
numpyArrayList2
```

```python
plt.plot(numpyArrayList, numpyArrayList2, "g--")
```

```python
plt.plot(numpyArrayList, numpyArrayList2, "r*-")
```

## subplot()

```python
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st graph
plt.plot(numpyArrayList, numpyArrayList2, "r*-")
plt.subplot(1, 2, 2)
plt.plot(numpyArrayList2, numpyArrayList, "--")
```

## figure()

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2, 0.2, 0.9, 0.9])
figureAxes.plot(numpyArrayList, numpyArrayList2, "g")
figureAxes.set_xlabel("X Axis")
figureAxes.set_ylabel("Y Axis")
figureAxes.set_title("Graphic")
```

# Nested Graphics

```python
figure2 = plt.figure()

axes1 = figure2.add_axes([0.1, 0.2, 0.9, 0.9])
axes2 = figure2.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(numpyArrayList, numpyArrayList2, "g")
axes1.set_xlabel("X Axis")
axes1.set_ylabel("Y Axis")
axes1.set_title("Main Graphic")

axes2.plot(numpyArrayList2, numpyArrayList, "r")
axes2.set_xlabel("X Axis")
axes2.set_ylabel("Y Axis")
axes2.set_title("Mini Graphic")
```

-----------------------

## subplots() and Nested Graphics

```python
(myFigure, myAxises)= plt.subplots(nrows=1, ncols=2)
type(myFigure)

for axis in myAxises:
    axis.plot(numpyArrayList, numpyArrayList2, "g")
    axis.set_xlabel("x axis")
    
#myFigure 
plt.tight_layout()
```

# Visual Enhancement

```python
newFigure = plt.figure(figsize = (5, 5))
# dpi = 150

newAxis = newFigure.add_axes([0.1, 0.1, 0.9, 0.9])

newAxis.plot(numpyArrayList, numpyArrayList ** 2, label = "numpyArray ** 2")
newAxis.plot(numpyArrayList, numpyArrayList ** 3, label = "numpyArray ** 3")

#indicator placement
newAxis.legend(loc = 5)
```

## savefig()

```python
# dpi = pixel
newFigure.savefig("myFigure.png", dpi = 200)
```

```python
numpyArr1 = np.linspace(0,10,20)
numpyArr2 = numpyArr1 ** 2

(myFigure2, myAxis2) = plt.subplots()
myAxis2.plot(numpyArr1, numpyArr2, color = "#b662b8", alpha = 0.3) #alpha = transpran
myAxis2.plot(numpyArr2, numpyArr1, color = "#c0e349")
```

```python
(myFigure3, myAxis3) = plt.subplots()
myAxis3.plot(numpyArr1, numpyArr1 + 2, color = "blue", linewidth = 1.0) 
myAxis3.plot(numpyArr1, numpyArr1 + 3, color = "red", linewidth = 1.0)

myAxis3.plot(numpyArr1, numpyArr1 + 4, color = "orange", linestyle = "--" )
myAxis3.plot(numpyArr1, numpyArr1 + 5, color = "purple", linestyle = ":" )
myAxis3.plot(numpyArr1, numpyArr1 + 6, color = "y", linestyle = "-." )
myAxis3.plot(numpyArr1, numpyArr1 + 7, color = "black", linestyle = "--", marker = "o", markersize = 6, markerfacecolor  = "red" )
myAxis3.plot(numpyArr1, numpyArr1 + 8, color = "black", linestyle = "--", marker = "+", markersize = 8 )
```
